### **Problem 75 ‚Äì Minimum Path Sum**

---

## üßÆ **Problem Statement:**

> Given a `m x n` grid filled with **non-negative numbers**, find a path from **top-left to bottom-right** which **minimizes the sum** of all numbers along the path.  
> You can only move **right** or **down**.

**Example:**
```text
Input: grid = [
  [1,3,1],
  [1,5,1],
  [4,2,1]
]
Output: 7  
Explanation: 1 ‚Üí 3 ‚Üí 1 ‚Üí 1 ‚Üí 1 = 7
```

---

## üåÄ **Algoverse Pathway Layers (L0 ‚Üí L6)**

| **Layer** | **Mapping in This Problem** |
|-----------|------------------------------|
| **L0: Primitives** | Grid of integers, positions `(i, j)` |
| **L1: Motifs** | Dynamic Programming, Min Accumulation |
| **L2: Structures** | In-place grid update or separate DP table |
| **L3: Complex Graphs** | DAG of directional movement (down/right) |
| **L4: Dynamics** | dp[i][j] = min(dp[i-1][j], dp[i][j-1]) + grid[i][j] |
| **L5: Meta-Cognition** | Reuse original grid to save space (if allowed) |
| **L6: Emergence** | Minimum cost path emerges bottom-up from choices |

---

## ‚úÖ Code (In-Place DP Optimization)

```python
def minPathSum(grid):
    """
    Agentic Simulation of Minimum Path Sum
    Agentome: Sense ‚Üí Memory ‚Üí Intellect ‚Üí Action ‚Üí Ego
    """

    rows, cols = len(grid), len(grid[0])

    # -------------------------------
    # üß† INTELLECT AGENT (L2‚ÄìL5)
    # -------------------------------
    for i in range(rows):
        for j in range(cols):
            if i == 0 and j == 0:
                continue  # start point
            elif i == 0:
                grid[i][j] += grid[i][j - 1]  # only from left
            elif j == 0:
                grid[i][j] += grid[i - 1][j]  # only from top
            else:
                grid[i][j] += min(grid[i - 1][j], grid[i][j - 1])

    # -------------------------------
    # üß† EGO AGENT (L6)
    # -------------------------------
    return grid[-1][-1]
```

- Choose the minimum and add to current cell

- At each cell, accumulate the min path sum so far

- Final answer is in the bottom-right cell: `grid[rows-1][cols-1]`
