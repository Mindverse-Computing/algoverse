## ðŸ§® **Problem Statement:**

> Given an integer array `nums`, return `true` if any value appears **at least twice**, and `false` if every element is **distinct**.

**Example:**
```text
Input: nums = [1,2,3,1]
Output: true

Input: nums = [1,2,3,4]
Output: false
```

## âœ… Code

```python
def contains_duplicate(nums):
    """
    Agentic Simulation of Contains Duplicate Problem
    Agentome: Sense â†’ Memory â†’ Intellect â†’ Action â†’ Ego
    """

    # -------------------------------
    # ðŸ§  SENSE AGENT (L0)
    # -------------------------------
    if not nums:
        return False

    # -------------------------------
    # ðŸ§  MEMORY AGENT (L1)
    # -------------------------------
    seen = set()

    # -------------------------------
    # ðŸ§  INTELLECT AGENT (L2-L5)
    # -------------------------------
    for num in nums:
        # -------------------------------
        # ðŸ§  META-COGNITION (L5)
        # -------------------------------
        if num in seen:
            return True
        # -------------------------------
        # ðŸ§  ACTION AGENT (L4)
        # -------------------------------
        seen.add(num)

    # -------------------------------
    # ðŸ§  EGO AGENT (L6)
    # -------------------------------
    return False
```
