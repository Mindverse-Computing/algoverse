## âœ… Code

```python
def max_product(nums):
    """
    Agentic Simulation of Maximum Product Subarray
    Agentome: Sense â†’ Memory â†’ Intellect â†’ Action â†’ Ego
    """

    # -------------------------------
    # ðŸ§  SENSE AGENT (L0)
    # -------------------------------
    if not nums:
        return 0

    # -------------------------------
    # ðŸ§  MEMORY AGENT (L1)
    # -------------------------------
    current_max = current_min = result = nums[0]

    # -------------------------------
    # ðŸ§  INTELLECT AGENT (L2-L5)
    # -------------------------------
    for num in nums[1:]:
        # âš ï¸ Store temp because current_max will be updated
        temp_max = max(num, num * current_max, num * current_min)
        current_min = min(num, num * current_max, num * current_min)
        current_max = temp_max

        # -------------------------------
        # ðŸ§  META-COGNITION (L5)
        # -------------------------------
        result = max(result, current_max)

    # -------------------------------
    # ðŸ§  EGO AGENT (L6)
    # -------------------------------
    return result
```
